## Concatenating Data Frame

```python
games_df = pd.read_csv('data/2017-2018_gamedata.csv')
```

```python
years = np.arange(2007,2018)
```

```python
games_df = pd.read_csv('data/2017-2018_gamedata.csv')
games_df['year']= 2018
for year in years:
    df = pd.read_csv('data/{}-{}_gamedata.csv'.format(year-1,year))
    df['year'] = year
    games_df = pd.concat((games_df,df),ignore_index=True)
```

```python
games_df
```

## Cleanup of data in games_df (box score dataframe)

```python
games_df['3PM-A'].fillna(games_df['3GM-A'], inplace=True)
```

```python
games_df.info()
```

```python
games_df[games_df['year']==2006]
```

```python
games_df.drop(['3GM-A','Unnamed: 0'], axis = 1,inplace=True)
```

```python
games_df.head()
```

Dividing strings for 3Ps FTs, FGs

```python
games_df.dropna(inplace=True)
```

```python
games_df['3PM'] = games_df['3PM-A'].str.split("-").str.get(0).astype(float)
games_df['3PA'] = games_df['3PM-A'].str.split("-").str.get(1).astype(float)
games_df['FGM'] = games_df['FGM-A'].str.split("-").str.get(0).astype(float)
games_df['FGA'] = games_df['FGM-A'].str.split("-").str.get(1).astype(float)
games_df['FTM'] = games_df['FTM-A'].str.split("-").str.get(0).astype(float)
games_df['FTA'] = games_df['FTM-A'].str.split("-").str.get(1).astype(float)
```

```python
games_df.drop(['3PM-A','FGM-A','FTM-A'], axis = 1,inplace=True)
```

```python
games_df['date'] = pd.to_datetime(games_df['DATE'], format='%Y-%m-%d')
```

Cleaning up rebounds

```python
games_df.rename(index = str, columns={'OFF':'OFF_REB', 'DEF':'DEF_REB'}, inplace=True)
```

```python
games_df.head()
```

Dividing home team and away team

```python
games_df['Away'] = games_df['MATCH'].str.split(" vs. ").str.get(0).str.strip('\n')
games_df['Home'] = games_df['MATCH'].str.split(" vs. ").str.get(1).str.strip('\n')
```

```python
games_df['TEAM'] = games_df['TEAM'].str.strip('\n')
```

```python
games_df.head()
```

Cleaning up minutes

```python
games_df['MINUTES']=games_df['MIN'].str.split(":").str.get(0).astype(float)
games_df.drop(['MIN'], axis = 1,inplace=True)
```

```python
games_df.head()
```

## Creating Game Outcomes dataframe

### Aggregating into matchups

```python
matchups = games_df.groupby(['year','DATE','date','MATCH','TEAM','Home','Away']).agg('sum').reset_index()
```

```python
matchups.head(50)
```

```python
matchups['home_team'] = (matchups['TEAM']==matchups['Home']).astype(int)
```

### Creating List of game outcomes

```python
df_temp = matchups.copy().sort_values(['date','MATCH','home_team']).reset_index()
```

```python
df_temp.head(50)
```

```python
matchups.head(50)
```

```python
df_temp.replace({'home_team': {0: -1}},inplace=True)
```

There seem to be matchups where there is not an away team.

```python
error_check = df_temp.groupby(['year','DATE','MATCH']).agg({'home_team':'sum'})
```

```python
error_check.groupby('year').agg('sum')
```

```python
#All games should have a value of 0 for home team, as I replaced 
error_check.rename(index = str, columns={'home_team':'h_sum'}, inplace=True)
error_check['h_sum'].sum()
```

There appear to be multiple games where there is only one team accounted for. Unsure why, but as they only represent a small fraction of the data (~3%) and are evenly spread across the years, I will drop them.

```python
game_outcomes = pd.merge(df_temp, error_check, how='left',on=['DATE','MATCH'])
```

```python
game_outcomes = game_outcomes[game_outcomes['h_sum']==0]
```

```python
game_outcomes['h_sum'].sum()
```

```python
game_outcomes.replace({'home_team': {-1: 0}},inplace=True)
```

```python
game_outcomes['home_point_spread'] = game_outcomes['PTS'].diff() * game_outcomes['home_team']
```

```python
game_outcomes.head()
```

```python
game_outcomes = game_outcomes.sort_values(['DATE','MATCH','home_team'])
```

```python
game_outcomes[['MATCH','TEAM','PTS','home_team','home_point_spread']].head()
```

```python
game_outcomes.drop(['h_sum','index'], axis=1, inplace=True)
```

```python
game_outcomes
```

```python
games_final = game_outcomes[['year','DATE','Home','Away','TEAM','home_point_spread']].reset_index()
```

```python
games_final
```

```python
games_final['home_winner'] = (games_final['home_point_spread']>0).astype(int)
```

```python
games_final = games_final[games_final['TEAM']==games_final['Home']]
```

```python
games_final.drop('TEAM', axis=1,inplace=True)
```
